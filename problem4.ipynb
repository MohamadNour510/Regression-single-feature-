Problem Statement:

You have a dataset containing the years of experience of employees and their monthly salary (in $1000s).
 
Your task is to implement Linear Regression using Gradient Descent to predict the salary based on years of experience.



```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

```

```python
#dataset
df=pd.read_csv("datasetforproblem4.csv")

X=df["Years_of_Experience"]

Y=df["Monthly_Salary"]


```

```python
#using feature scalling by z_score normalization

X_Norm=(X-np.mean(X))/np.std(X)
X_Norm
```

```python
# Initialize parameters


theta0 = 0  # Intercept
theta1 = 0  # Slope
alpha = 0.01  # Learning rate
iterations = 1000  # Number of iterations
m = len(X)  # Number of data points

 
```

```python
#Apply gradient decsent Algorithm
cost_history=[]

for _ in range(iterations):
    y_pred=theta0+theta1*X_Norm
    
    #compute cost (error square)
    cost=((1/2*m)*np.sum(y_pred-Y)**2)
    cost_history.append(cost)
    #compute gradient
    d_theta0=(1/m)*np.sum(y_pred-Y)
    
    d_theta1=(1/m)*np.sum((y_pred-Y)*X_Norm)
    
    #update parameters
    theta0-=alpha*d_theta0
    theta1-=alpha*d_theta1
    

# Print final parameters
print(f"Final theta0 (Intercept): {theta0}")
print(f"Final theta1 (Slope): {theta1}")

    

```

```python
# Plot cost function vs. iterations

plt.plot(range(iterations),cost_history,label="cost")

plt.xlabel("iterations")
plt.ylabel("Cost")
plt.title("Plot cost function vs. iterations")
plt.legend()
plt.show()
```

```python
plt.scatter(X,Y,color="blue",label="Acual Value")
plt.plot(X,theta0+theta1*((X-np.mean(X))/np.std(X)),color="red",label="Regression Line")

plt.xlabel("Years_of_Experience")
plt.ylabel("Monthly_Salary in 1000$")

plt.title("Years_of_Experience vs Monthly_Salary")

plt.legend()

plt.show()

```

```python
x_test=(7-np.mean(X))/np.std(X)

predicted_salary=theta0+theta1*x_test

print(f"Monthly Salary for Year of expereince: ${predicted_salary*1000:.2f}")
```
