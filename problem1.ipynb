Problem Statement:

You are given a dataset that contains information about house sizes (in square meters) and their corresponding prices (in $1000). Your task is to implement Linear Regression from scratch using Gradient Descent to minimize the Mean Squared Error (MSE) cost function.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X = np.array([50, 60, 75, 90, 120, 150] )
Y = np.array([150, 180, 210, 250, 300, 360])

	
```

```python
# Normalize feature for better convergence (optional)
X_mean = np.mean(X)
X_std = np.std(X)
X_norm = (X - X_mean) / X_std  # Standardization
```

```python
# Initialize parameters (theta0 = intercept, theta1 = slope)
theta0 = 0  # Bias term
theta1 = 0  # Weight term
alpha = 0.01  # Learning rate
iterations = 1000  # Number of iterations
m = len(X)  # Number of data points
```

```python
# Cost function history for plotting
cost_history = []

# Gradient Descent Algorithm
for _ in range(iterations):
    # Compute predictions
    y_pred = theta0 + theta1 * X_norm
    
    # Compute cost (Mean Squared Error)
    cost = (1 / (2 * m)) * np.sum((y_pred - Y) ** 2)
    cost_history.append(cost)
    
    # Compute gradients
    d_theta0 = (1 / m) * np.sum(y_pred - Y)
    d_theta1 = (1 / m) * np.sum((y_pred - Y) * X_norm)
    
    # Update parameters
    theta0 -= alpha * d_theta0
    theta1 -= alpha * d_theta1


# Print final parameters
print(f"Final theta0 (Intercept): {theta0}")
print(f"Final theta1 (Slope): {theta1}")

```

```python
# Plot cost function history
plt.plot(range(iterations), cost_history, label="Cost")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Function Convergence")
plt.legend()
plt.show()
```

```python
# Visualizing the Best-Fit Line
plt.scatter(X, Y, color='red', label="Actual Data")
plt.plot(X, theta0 + theta1 * ((X - X_mean) / X_std), color='blue', label="Regression Line")
plt.xlabel("House Size (m²)")
plt.ylabel("Price ($1000)")
plt.title("House Size vs Price Regression")
plt.legend()
plt.show()
```

```python
# Predict the price for a house of size 100m²
X_test = (100 - X_mean) / X_std  # Normalize test input
predicted_price = theta0 + theta1 * X_test
print(f"Predicted price for a 100m² house: ${predicted_price*1000:.2f}")
```
