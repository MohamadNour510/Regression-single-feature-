Dataset (Years Since Built ‚Üí House Price in $1000s)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Years Since Built (X) and House Price in $1000s (Y)
X = np.array([1, 5, 10, 15, 20, 25, 30, 35, 40])
Y = np.array([250, 220, 180, 150, 130, 115, 100, 90, 75])
```

```python
#feature scalling 

x_mean=np.mean(X)

x_std=np.std(X)

X_Norm=(X-x_mean)/x_std

```

```python
# Gradient Descent Initialization
theta0 = 0  # Intercept
theta1 = 0  # Slope
alpha = 0.1  # Learning rate
iterations = 1000  # Number of iterations
m = len(X)  # Number of training examples
cost_history = []
```

```python
#Gradient Descent Algorithm

for _ in range(iterations):
    y_pred=theta0+theta1*X_Norm #predict value
    
    #compute cost (error square)
    cost=((1/2*m)*np.sum(y_pred-Y)**2)
    cost_history.append(cost)
    #compute gradient
    d_theta0=(1/m)*np.sum(y_pred-Y)
    
    d_theta1=(1/m)*np.sum((y_pred-Y)*X_Norm)
    
    #update parameters
    theta0-=alpha*d_theta0
    theta1-=alpha*d_theta1
    
# Print final parameters
print(f"Final theta0 (Intercept): {theta0}")
print(f"Final theta1 (Slope): {theta1}")

     
    
```

```python
len(cost_history)
```

```python
# Plot cost function vs. iterations

plt.plot(range(iterations),cost_history,label="cost")

plt.xlabel("iterations")
plt.ylabel("Cost")
plt.title("Plot cost function vs. iterations")
plt.legend()
plt.show()
```

```python
plt.scatter(X,Y,color="blue",label="Acual Value")
plt.plot(X,theta0+theta1*((X-np.mean(X))/np.std(X)),color="red",label="Regression Line")

plt.xlabel("Years_Since_built")
plt.ylabel("House price in 1000$")

plt.title("Years_of_Experience vs Monthly_Salary")

plt.legend()

plt.show()
```

```python
x_test=(7-x_mean)/x_std

predicted_house_price=theta0+theta1*x_test

print(f"Houese price:{predicted_house_price*1000:.2f}$ ")
```

```python

# Using Scikit-learn Linear Regression
X_Reshape=X.reshape(-1,1)
model=LinearRegression()

model.fit(X_Reshape,Y)

predictedHousePriceUsingSklearn=model.predict([[7]])[0]

print(f"Houese price:{predictedHousePriceUsingSklearn*1000:.2f}$ ")
```
