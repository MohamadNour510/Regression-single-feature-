```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Exercise: Predict Student Scores Using Linear Regression
Problem Statement:

A studentâ€™s performance on a final exam is influenced by the number of hours they study. You are given a dataset where each data point consists of the number of hours a student studied and their corresponding exam score.

Your task is to implement Linear Regression from scratch using Gradient Descent to predict a student's score based on their study hours.

1-Load the dataset

Store the study hours in an array X


Store the corresponding exam scores in an array Y.

```python
# Given dataset (Study Hours vs Exam Scores)
X = np.array([1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0])
Y = np.array([50, 55, 60, 65, 70, 75, 80, 85, 88, 90])

```

```python

# Normalize features for better convergence
X_Norm = (X - np.mean(X)) / np.std(X)
X_Norm
```

```python
# Initialize parameters
theta0 = 0  # Intercept
theta1 = 0  # Slope
alpha = 0.1  # Learning rate
iterations = 1000  # Number of iterations
m = len(X)  # Number of data points


```

```python

# Clear cost history before running gradient descent
cost_history = []  

# Apply Gradient Descent
for _ in range(iterations):
    y_pred = theta0 + theta1 * X_Norm  # Hypothesis function

    # Compute cost (Mean Squared Error)
    cost = (1 / (2 * m)) * np.sum((y_pred - Y) ** 2)
    cost_history.append(cost)

    # Compute gradients
    d_theta0 = (1 / m) * np.sum(y_pred - Y)
    d_theta1 = (1 / m) * np.sum((y_pred - Y) * X_Norm)

    # Update parameters
    theta0 -= alpha * d_theta0
    theta1 -= alpha * d_theta1

# Print final parameters
print(f"Final theta0 (Intercept): {theta0}")
print(f"Final theta1 (Slope): {theta1}")
    
    

```

```python
# Plot cost function vs. iterations
plt.plot(range(iterations), cost_history, label="Cost Function")
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.title("Gradient Descent Convergence")
plt.legend()
plt.show()
```

```python
plt.scatter(X,Y,color="red",label="Acual Value")
plt.plot(X,theta0+theta1 * ((X - np.mean(X)) / np.std(X)),color="blue",label="Regression Line")
plt.xlabel("The Numer of Hours")
plt.ylabel("Student Score")
plt.title("Student Final Score Depend on the number of  hours")
plt.legend()
plt.show()
```

```python
# Predict Exam Score for 4.7 Hours of Study
X_test = (4.7 - np.mean(X)) / np.std(X)  # Normalize input
predicted_score = theta0 + theta1 * X_test
print(f"Predicted Exam Score for 4.7 Hours of Study: {predicted_score:.2f}%")
```
