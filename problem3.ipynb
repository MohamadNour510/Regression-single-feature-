Problem Statement:

You are given a dataset containing the age of cars (in years) and their market price (in $1000s). 

Your goal is to implement Linear Regression using Gradient Descent to predict the price of a car based on its age.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Given dataset (Age Car  Hours vs Prices)
X=np.array([1,2,3,4,5,6,7,8])
Y=np.array([50,45,40,36,30,27,22,20])
```

```python
#Feature Scalling (Normalization)

x_norm=(X-np.mean(X))/np.std(X)

x_norm
```

```python
# Initialize parameters
theta0 = 0  # Intercept
theta1 = 0  # Slope
alpha = 0.1  # Learning rate
iterations = 1000  # Number of iterations
m = len(X)  # Number of data points




```

```python
#Apply Gradient Descent Algorithm
cost_history=[]
for _ in range(iterations):
    y_pred=theta0+theta1*x_norm
    
    # Compute cost (Mean Squared Error)
    cost=((1/2*m)*np.sum(y_pred-Y)**2)
    cost_history.append(cost)
    
    # Compute gradients
    d_theta0 = (1 / m) * np.sum(y_pred - Y)
    d_theta1 = (1 / m) * np.sum((y_pred - Y) * x_norm)
    
    
    # Update parameters
    
    theta0-=alpha*d_theta0
    theta1-=alpha*d_theta1
    
# Print final parameters
print(f"Final theta0 (Intercept): {theta0}")
print(f"Final theta1 (Slope): {theta1}")
cost
```

```python
# Plot cost function vs. iterations

plt.plot(range(iterations),cost_history,label="Cost Function")

plt.xlabel("iterations")
plt.ylabel("Cost")

plt.title("Cost Function")

plt.legend()
plt.show()
```

```python

plt.scatter(X,Y,color="blue",label="Acual Value")
plt.plot(X,theta0+theta1*((X-np.mean(X))/np.std(X)),color="red",label="Regression Line")

plt.xlabel("Age Car")
plt.ylabel("Price in 1000$")

plt.title("Age car vs its price")

plt.legend()

plt.show()


            
```

```python
x_test=(5.5-np.mean(X))/np.std(X)

predicted_price=theta0+theta1*x_test

print(f"Predicted price for a Age car: ${predicted_price*1000:.2f}")
```
